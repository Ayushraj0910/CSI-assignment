## 6) Basic cleaning

Handle missing values, remove duplicates, and make sure numeric columns are clean.

```python
df = df.copy()
df.columns = [c.strip().lower() for c in df.columns]
df.drop_duplicates(inplace=True)

numeric_cols = [c for c in df.columns if c != 'country']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

medians = df[numeric_cols].median()
df[numeric_cols] = df[numeric_cols].fillna(medians)

df.isna().sum()
```

## 7) Exploratory Data Analysis

```python
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
for i, col in enumerate(numeric_cols):
    plt.figure(figsize=(8, 3))
    sns.boxplot(x=df[col])
    plt.title('Boxplot: ' + col)
    plt.tight_layout()
    plt.show()
```

## 8) Feature scaling

Clustering works better when features are on the same scale.

```python
features = df[numeric_cols]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
X_scaled.shape
```

## 9) K-Means: Elbow method

```python
k_values = list(range(2, 11))
inertias = []

for k in k_values:
    inertias.append(KMeans(n_clusters=k, random_state=42, n_init=10).fit(X_scaled).inertia_)

plt.figure(figsize=(8, 4))
plt.plot(k_values, inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

## 10) Train K-Means

```python
best_k = 3  # change after checking the elbow plot
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled)
df['kmeans_cluster'] = cluster_labels

sil_score = silhouette_score(X_scaled, df['kmeans_cluster'])
print('Silhouette Score:', sil_score)

if 'country' in df.columns:
    df[['country', 'kmeans_cluster']].head()
else:
    df[['kmeans_cluster']].head()
```

## 11) Try DBSCAN

```python
dbscan = DBSCAN(eps=1.5, min_samples=5)
df['dbscan_cluster'] = dbscan.fit_predict(X_scaled)

dbscan_counts = df['dbscan_cluster'].value_counts().sort_index()
print(dbscan_counts)
```

## 12) PCA visualization

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

viz = pd.DataFrame(X_pca, columns=['pca1', 'pca2'])
viz['cluster'] = df['kmeans_cluster'].values

plt.figure(figsize=(8, 6))
sns.scatterplot(data=viz, x='pca1', y='pca2', hue='cluster', palette='tab10')
plt.title('K-Means Clusters Visualized with PCA')
plt.show()
```

## 14) Final insights

Results below are from running this notebook end-to-end on the Kaggle **Country-data.csv** dataset (167 countries, `best_k = 3`, silhouette score ≈ 0.283).

**Cluster 0 — Developed / high-income (36 countries).** This group has the lowest child mortality (around 5.0 per 1,000 births) and the highest life expectancy (about 80 years). Average income (~$45,700) and gdpp (~$42,500) are clearly the highest among the three groups, and fertility is the lowest (~1.75). Countries like Australia, Austria, Belgium and Canada fall here — basically stable, wealthy economies.

**Cluster 1 — Underdeveloped / needs aid (47 countries).** Child mortality shoots up to roughly 93 per 1,000, life expectancy drops to about 59 years, and both income (~$3,940) and gdpp (~$1,920) are the lowest of all clusters. Inflation is also the highest here (~12%) along with fertility (~5 children per woman). Afghanistan, Angola, Benin, Burkina Faso and Burundi are examples. **These countries should be first in line for aid**, since poor health outcomes are paired with the least economic ability to fix them.

**Cluster 2 — Developing / middle income (84 countries, the biggest bucket).** Every metric here sits between the other two — child mortality around 22, life expectancy near 73 years, income close to $12,300 and gdpp around $6,490. Albania, Algeria, Argentina and Bangladesh land in this cluster. They're transitioning economies, could use some targeted support, but aren't in a crisis like Cluster 1.

**About the clustering quality:** a silhouette score of ~0.28 means the clusters are reasonably separated but do overlap a bit, which makes sense for real-world socio-economic data instead of neat synthetic blobs. The PCA plot (first two components cover about 63% of the variance) shows three groups that are mostly distinguishable with a bit of mixing at the edges — matching the silhouette score. DBSCAN with `eps=1.5` isn't very useful here since it dumps almost everything into one big cluster plus some noise points, so for this dataset K-Means gives a cleaner 3-way split. DBSCAN would probably need a smaller `eps` value to pick up similar structure.
