### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)  

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]
    Av = A @ v
    lambda_v = lam * v
    print(f"Eigenpair {i+1}: A@v = {Av}, λ*v = {lambda_v}, match: {np.allclose(Av, lambda_v)}")
    assert np.allclose(Av, lambda_v), f"Eigenvector equation failed for pair {i+1}"

# Plot
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]
    lam_v = lam * v
    plt.quiver(0, 0, lam_v[0], lam_v[1], angles='xy', scale_units='xy', scale=1, color=f'C{i}', label=f'λ={lam:.2f}')

plt.xlim(-4, 8)
plt.ylim(-1, 8)
plt.axhline(0, color='black', lw=0.4)
plt.axvline(0, color='red', lw=0.4)
plt.gca().set_aspect('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Eigenvectors scaled by eigenvalues')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

**Your Geometric Explanation (4.3):**

Eigenvectors are special vectors that, when multiplied by a matrix, change only in length — they may stretch or shrink — without changing direction. The one exception is when the eigenvalue is negative: in that case, the vector flips by 180°, but it's still considered to lie along the same line.

Eigenvalues tell us how much the corresponding eigenvector is scaled:
- If λ > 1, the vector is stretched (gets longer).
- If 0 < λ < 1, the vector is compressed/squished (gets shorter).
- If λ = 1, the vector remains unchanged.
- If λ = 0, the vector is mapped to the zero vector.
- If λ < 0, the vector flips by 180° and is scaled by |λ|.

When a matrix stretches a vector, it increases the vector's magnitude. When it squishes or compresses a vector, it decreases the magnitude. For an eigenvector specifically, this change in length happens without any change in direction — aside from the possible 180° flip when the eigenvalue is negative.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)

X_reconstructed = U @ np.diag(S) @ Vt
X_approx        = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

@ what are U, S, Vt in SVD, and how do they relate to principal components? Which matrix contains the directions of maximum variance?

#When we decompose a matrix X using SVD, wwe get X = U @ S @ Vt where 

**U** determines the left singular vectors as its columns , it describe's the relationship between the rows of the original data.

**S** is the list of singular values, stored in the form of largest to smallest. Each one tells us that how important or strong the corresponding direction is ie the first singular value captures the most significant pattern in the data, the second captures the next most significant, and so on.

The matrix **Vt** contains the singular vectors in its rows. They tell us the directions in the feature space; in other words, what combinations of your initial columns are important. That’s the link to PCA’s principal components.

**SVD** and **PCA** share many similarities, and PCA usually uses SVD for its computations. When you center your data by subtracting the mean of each column from the data and perform SVD on this centered matrix, the rows of the transpose of V would give the principal components, while the singular values in S would inform you about how much variance the principal components explain. U @ S would give the data expressed on principal component axes. In other words, centering the data and applying SVD would be equivalent to PCA.

The matrix **Vt** holds the directions of the highest variances. The rows of **Vt** are all these directions, which are arranged in the decreasing order of the amount of variance captured by them, corresponding to the order of the singular values in **S**.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    hist_expected, bin_edges = np.histogram(expected, bins=bins, density=True)
    hist_actual, _ = np.histogram(actual, bins=bin_edges, density=True)

    expected_pct = hist_expected / np.sum(hist_expected)
    actual_pct = hist_actual / np.sum(hist_actual)

    epsilon = 1e-10
    expected_pct += epsilon
    actual_pct += epsilon

    psi = np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))
    return psi

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
ax.hist(train_dist, bins=30, alpha=0.5, label='Training' ,color='green')
ax.hist(drift_dist, bins=30, alpha=0.5, label='Production' ,color='purple' \
'')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** the underlying relationship between features and the target shifts over time, so the model's predictions become less accurate even though the input feature distributions haven't changed.
- **Covariate drift:** the distribution of input features shifts over time, while the relationship between features and the target remains roughly unchanged.
- **PSI < 0.1:** little to no shift so the distributions can be considered stable.
- **PSI 0.1–0.2:** moderate shift is observed thus worth monitoring and checking whether retraining is necessary.
- **PSI > 0.2:** major shift can be seen that usually signals a meaningful change in the data and typically calls for retraining or a model review.

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).

   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = red / total
p_blue  = blue / total
p_green = green / total

# Joint: P(red first AND blue second)
p_red_then_blue = (red / total) * (blue / (total - 1))

# Conditional: P(blue second | red first)
p_blue_given_red = p_red_then_blue / p_red

# Independence check: are P(B|A) == P(B) ?
independent = p_blue_given_red == p_blue  # YOUR CODE HERE — True/False

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_nspam  = 0.05
p_nspam             = 1 - p_spam

# P(free) — law of total probability
p_free = p_free_given_spam * p_spam + p_free_given_nspam * p_nspam

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_nspam):
    prior_nspam = 1 - prior_spam
    p_word = p_word_given_spam * prior_spam + p_word_given_nspam * prior_nspam
    posterior_spam = (p_word_given_spam * prior_spam) / p_word
    return posterior_spam

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | The probability that an email is spam, before considering any of its words |
| Likelihood | P(word \| Spam) | The probability that the word "free" appears, given that the email is spam |
| Evidence | P(word) | The probability that the word "free" shows up in an email overall |
| Posterior | P(Spam \| word) | The probability that an email is spam, given that it contains the word "free" |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = [np.mean(np.random.choice(population, size=sample_size, replace=True)) for _ in range(n_samples)]
sample_means = np.array(sample_means) 

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
ax.hist(sample_means, bins=30, density=True, alpha=0.7, label='Sample Means Histogram')
x = np.linspace(min(sample_means), max(sample_means), 300)
ax.plot(x, stats.norm.pdf(x, loc=pop_mean, scale=clt_std), color='red', label='Normal PDF (CLT)')
ax.set_xlabel('Sample Mean');
ax.set_ylabel('Density');
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, stats.norm(loc=pop_mean, scale=clt_std).cdf)
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

According to the Central Limit Theorem, as sample size grows, the distribution of sample means tends toward normality — regardless of what shape the original population follows. This is precisely why many classical statistical methods, such as t-tests, confidence intervals, ANOVA, and regression, are able to rely on a normality assumption in the first place.
